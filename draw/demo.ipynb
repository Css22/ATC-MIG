```python
import matplotlib.pyplot as plt

# 数据
labels = ['MISO', 'OUR']
average_queue_time = [284252.659, 62594.068]
JCT = [287282.678, 81361.07]
makespan = [589415, 178230]

# 绘制柱状图
x = range(len(labels))

plt.bar(x, average_queue_time, width=0.2, label='Average Queue Time')
plt.bar([i + 0.2 for i in x], JCT, width=0.2, label='JCT')
plt.bar([i + 0.4 for i in x], makespan, width=0.2, label='Makespan')

# 设置 x 轴标签和标题
plt.xlabel('Cases')
plt.xticks([i + 0.2 for i in x], labels)
plt.title('32')

# 设置图例
plt.legend()

# 显示图表
plt.show()
```
