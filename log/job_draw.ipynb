```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 准备自己的数据集
data = {
    'Category': ['A', 'B', 'C', 'A', 'B', 'C'],
    'Value': [10, 15, 8, 12, 9, 11],
    'Group': ['Group1', 'Group2', 'Group1', 'Group2', 'Group1', 'Group2']
}
df = pd.DataFrame(data)

# 设置子图布局
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

# 生成三个子图
for i, group in enumerate(df['Group'].unique()):
    ax = axes[i]
    subset = df[df['Group'] == group]
    sns.barplot(data=subset, x='Category', y='Value', ax=ax)
    ax.set_title(f'Group: {group}')
    ax.set_xlabel('Category')
    ax.set_ylabel('Value')

# 调整子图布局
plt.tight_layout()

# 显示图形
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
ax.yaxis.grid(color='gray', linestyle='dashed')
def addlabels(x,y,ax):
    for i in range(len(x)):
        ax.text(i, y[i] + 1, y[i], ha = 'center')

BIGGER_SIZE = 15.5

plt.rc('font', size=BIGGER_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=BIGGER_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=BIGGER_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

fig, (ax, ax2, ax3) = plt.subplots(1, 3)
fig.set_figheight(7)
fig.set_figwidth(21)

x = ["a", "b", "c", "d"]
y = [100, 60, 13.7, 7.6]

s = pd.Series(
    y,
    index = x,
)
ax.set_ylabel('Percentage')
ax.set_xlabel('A')
ax.set_ylim([0, 100])
addlabels(x, y,ax)
ax.tick_params(axis='x', colors='black')
ax.tick_params(axis='y', colors='black')
my_colors = ('#83d0de', '#adddd4', '#f3f165', '#ffb654', '#f39696')  #red, green, blue, black, etc.
my_colors2 = ('#298090', '#4dbba5', '#cdb85c', '#c19458', '#ab4848')  #red, green, blue, black, etc

handles = []
for i in range(4):
    handles.append(mpatches.Patch(color=my_colors[i], edgecolor=my_colors2[i], label=x[i]))

s.plot( 
    kind='bar', 
    color=my_colors,
    edgecolor = my_colors2,
    xticks=[], 
    width=1.0,
    ax=ax
)
y2 = [0.4, 1, 3, 5]

ax2.set_ylim([0, 10])
ax2.set_ylabel('Temperature')
ax2.set_xlabel('B')
addlabels(x, y2,ax2)
s2 = pd.Series(
    y2,
    index = x,
)

s2.plot( 
    kind='bar', 
    color=my_colors,
    edgecolor = my_colors2,
    xticks=[], 
    width=1.0,
    ax=ax2
)

y3 = [10, 20, 50, 78]

ax3.set_ylim([0, 100])
ax3.set_ylabel('Score')
ax3.set_xlabel('C')
addlabels(x, y3, ax3)
s3 = pd.Series(
    y3,
    index = x,
)
s3.plot( 
    kind='bar', 
    color=my_colors,
    edgecolor = my_colors2,
    xticks=[], 
    width=1.0,
    ax=ax3
)

lgd = fig.legend(handles=handles,bbox_to_anchor=(0.65, 1.0), ncol = 4)
fig.savefig('case_plot.png', bbox_extra_artists=(lgd,), bbox_inches='tight')
```
